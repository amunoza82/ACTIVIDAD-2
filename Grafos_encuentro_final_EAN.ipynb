```python
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_edge("A","B",weight= 1.0)
G.add_edge("A","C",weight= 1.0)
G.add_edge("A","D",weight= 1.0)
G.add_edge("A","E",weight= 1.0)
G.add_edge("A","F",weight= 1.0)
G.add_edge("B","A",weight= 1.0)
G.add_edge("B","D",weight= 1.0)
G.add_edge("B","E",weight= 1.0)
G.add_edge("B","F",weight= 1.0)
G.add_edge("C","A",weight= 1.0)
G.add_edge("C","D",weight= 1.0)
G.add_edge("C","E",weight= 1.0)
G.add_edge("C","F",weight= 1.0)
G.add_edge("D","A",weight= 1.0)
G.add_edge("D","B",weight= 1.0)
G.add_edge("D","C",weight= 1.0)
G.add_edge("D","F",weight= 1.0)
G.add_edge("E","A",weight= 1.0)
G.add_edge("E","B",weight= 1.0)
G.add_edge("E","C",weight= 1.0)
G.add_edge("E","F",weight= 1.0)
G.add_edge("F","A",weight= 1.0)
G.add_edge("F","B",weight= 1.0)
G.add_edge("F","C",weight= 1.0)
G.add_edge("F","D",weight= 1.0)
G.add_edge("F","E",weight= 1.0)

```

```python
import matplotlib.pyplot as plt
layout=nx.spring_layout(G)
nx.draw_circular(G,with_labels=True, node_color='#22bbbb',node_size=500)
plt.axis("equal")
plt.show()

```
